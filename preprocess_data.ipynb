```python
import pandas as pd

# 1. Create a list of all the season strings.
seasons = [f"season-{i}{i+1}" for i in range(9, 20)]  # Generates season-0910, season-1011, ..., season-1910

dataframes = []

for season in seasons:
    # 2. Construct filename based on season string.
    filename = f'spanish-la-liga_zip/data/{season}_csv.csv'
    
    try:
        # 3. Read the CSV file into a dataframe.
        temp_df = pd.read_csv(filename)
        temp_df['season'] = season 
        
        # 4. Append the dataframe to the list.
        dataframes.append(temp_df)
    except FileNotFoundError:
        print(f"File for {season} not found!")

# 5. Concatenate all dataframes in the list.
df = pd.concat(dataframes, ignore_index=True)
```

```python
columns_to_keep = ['season','HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'HTHG', 'HTAG', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC']
df = df[columns_to_keep]
```

```python
df.rename(columns={
    'FTHG': 'FullTime_HomeGoals',
    'FTAG': 'FullTime_AwayGoals',
    'HTHG': 'HalfTime_HomeGoals',
    'HTAG': 'HalfTime_AwayGoals',
    'HST': 'Home_ShotsOnTarget',
    'AST': 'Away_ShotsOnTarget',
    'HF': 'Home_Fouls',
    'AF': 'Away_Fouls',
    'HC': 'Home_Corners',
    'AC': 'Away_Corners'
}, inplace=True)
```
